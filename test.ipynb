# Output examples

```python
from ThickMNIST import ThickMNIST
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline

path_to_mnist = 'data/mnist'
```

## Example with dilation ratio of 1

```python
dataloader = ThickMNIST('train', path_to_mnist, dilation_ratio=1)
dataloader_original = ThickMNIST('train', path_to_mnist, dilation_ratio=0)

x_all = []
x_all_no_dilation = []
for i in [1, 3, 5, 7, 2, 0, 13, 15, 17, 4]:
    x_all.append(dataloader[i][0].numpy().squeeze())
    x_all_no_dilation.append(dataloader_original[i][0].numpy().squeeze())
x_all = np.hstack(x_all)
x_all_no_dilation = np.hstack(x_all_no_dilation)


plt.imshow(x_all, cmap='gray')
plt.title('Dilated')
plt.show()

plt.imshow(x_all_no_dilation, cmap='gray')
plt.title('Original')
plt.show()
```
